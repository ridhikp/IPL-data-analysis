```python
# Data handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
sns.set(style='whitegrid')
```

```python
# Load CSV files from local folder
matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')
most_runs = pd.read_csv('most_runs_average_strikerate.csv')
team_home_away = pd.read_csv('teamwise_home_and_away.csv')
teams = pd.read_csv('teams.csv')

# Load Excel file
players = pd.read_excel('Players.xlsx')
```

```python
# preview each dataset
print("Matches")
display(matches.head())

print("\nDeliveries")
display(deliveries.head())

print("\nMost Runs")
display(most_runs.head())

print("\nTeamwise Home and Away")
display(team_home_away.head())

print("\nTeams")
display(teams.head())

print("\nPlayers")
display(players.head())
```

```python

# Check for null values in each dataset
print("Matches:")
print(matches.isnull().sum(), '\n')

print("Deliveries:")
print(deliveries.isnull().sum(), '\n')

print("Most Runs:")
print(most_runs.isnull().sum(), '\n')

print("Teamwise Home and Away:")
print(team_home_away.isnull().sum(), '\n')

print("Teams:")
print(teams.isnull().sum(), '\n')

print("Players:")
print(players.isnull().sum())

```

```python
if 'umpire3' in matches.columns:
    matches.drop('umpire3', axis=1, inplace=True)
```

```python
# Matches cleaning
matches['city'].fillna('Unknown', inplace=True)

if 'umpire3' in matches.columns:
    matches.drop('umpire3', axis=1, inplace=True)

# Most runs dataset
most_runs['average'].replace('', np.nan, inplace=True)

# Players dataset
players['Batting_Hand'].fillna('Unknown', inplace=True)
```

```python
# analyzing the matches dataset to get insights using EDA
# To check how many IPL matches were played each season

#Number of matches per season
matches_per_season = matches['Season'].value_counts().sort_index()


plt.figure(figsize=(10, 5))
sns.barplot(x=matches_per_season.index, y=matches_per_season.values, palette='viridis')
plt.title('Number of Matches Per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#to find most successful teams i.e. total wins
#Total wins per team
wins = matches['winner'].value_counts()

plt.figure(figsize=(10,5))
sns.barplot(x=wins.values, y=wins.index, palette= 'magma')
plt.title('Most successful Teams in IPL(by Wins)')
plt.xlabel('wins')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```

```python
#TO find out whether teams prefer to bat or field first after winning the toss
toss_decision = matches['toss_decision'].value_counts()

plt.figure(figsize=(6, 4))
sns.barplot(x=toss_decision.index, y=toss_decision.values, palette='coolwarm')
plt.title('Toss Decision Counts')
plt.xlabel('Decision')
plt.ylabel('count')
plt.tight_layout()
plt.show()
```

```python
#To check how often the team that wins toss also wins the match.
toss_match_win = matches[matches['toss_winner'] == matches['winner']]
win_percentage = (len(toss_match_win) / len(matches)) * 100

print(f"âœ… Toss winner also won the match in {win_percentage:.2f}% of games.")
```

```python
# Player performance analysis
most_runs.head()
```

```python
#Top 10 Run Scorers
top_scorers = most_runs.sort_values(by='total_runs', ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(y=top_scorers['batsman'], x=top_scorers['total_runs'], palette='crest')
plt.title('Top 10 Run Scorers in IPL')
plt.xlabel('Total Runs')
plt.ylabel('Batsman')
plt.tight_layout()
plt.show()
```

```python
# Top 10 by Strike Rate 
# To findout which batsmen score fastest it filters out players with very few matches
most_runs_filtered = most_runs[most_runs['total_runs'] >= 500]

top_strikers = most_runs_filtered.sort_values(by= 'strikerate', ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_strikers['strikerate'], y=top_strikers['batsman'], palette='flare')
plt.title('Top 10 Strike Rate')
plt.xlabel('strike rate')
plt.ylabel('Batsman')
plt.tight_layout()
plt.show()
```

```python
# TO find out top 10 batting average
#to remove not out text values if present
most_runs_filtered = most_runs_filtered[most_runs_filtered['average'] != 'Not Out']
#convert the average to nemeric
most_runs_filtered['average'] = pd.to_numeric(most_runs_filtered['average'])

top_averages = most_runs_filtered.sort_values(by= 'average', ascending=False).head(10)

#visualization
plt.figure(figsize=(10, 5))
sns.barplot(x=top_averages['average'], y=top_averages['batsman'], palette= 'Blues_r')
plt.title('Top 10 Batting Average')
plt.xlabel('average')
plt.ylabel('batsman')
plt.tight_layout()
plt.show()
```

```python
#Balance Between Strike Rate and Average
# to find who is fast 
plt.figure(figsize=(10, 6))
sns.scatterplot(data=most_runs_filtered, x='strikerate', y='average', hue='total_runs', palette='viridis', size='total_runs', sizes=(50, 300))
plt.title('Batting Average vs Strike Rate (Size = Total Runs)')
plt.xlabel('Strike Rate')
plt.ylabel('Average')
plt.tight_layout()
plt.show()
```

```python
#Bowling analysis
deliveries.head()
```

```python

# Filter only valid dismissals exclude run outs, retired hurt, etc.
valid_dismissals = deliveries[~deliveries['dismissal_kind'].isin(['run out', 'retired hurt', 'obstructing the field'])]

# Drop rows where no player was dismissed
valid_dismissals = valid_dismissals[valid_dismissals['player_dismissed'].notna()]

# Count wickets by bowler
top_wicket_takers = valid_dismissals['bowler'].value_counts().head(10)

# Plot
plt.figure(figsize=(10,5))
sns.barplot(x=top_wicket_takers.values, y=top_wicket_takers.index, palette='inferno')
plt.title('Top 10 Wicket-Takers in IPL')
plt.xlabel('Wickets')
plt.ylabel('Bowler')
plt.tight_layout()
plt.show()
```

```python
# to find bowlers who take most wickets in overs 16 to 20
#top death over bowlers
death_overs = valid_dismissals[valid_dismissals['over'] >= 16]
death_wickets = death_overs['bowler'].value_counts().head(10)
# Plot
plt.figure(figsize=(10,5))
sns.barplot(x=death_wickets.values, y=death_wickets.index, palette='rocket')
plt.title('Top 10 Wicket-Takers in Death Overs (16â€“20)')
plt.xlabel('Wickets')
plt.ylabel('Bowler')
plt.tight_layout()
plt.show()
```

```python
#team hone vs away analysis
home_away = pd.read_csv('teamwise_home_and_away.csv')
home_away.head()
```

```python
#comparing it side by side
# Sort teams by home win %
home_away = home_away.sort_values(by='home_win_percentage', ascending=False)

# Plot side-by-side bars
plt.figure(figsize=(10,6))
plt.bar(home_away['team'], home_away['home_win_percentage'], label='Home Win %', color='lightblue')
plt.bar(home_away['team'], home_away['away_win_percentage'], label='Away Win %', color='yellow', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Win Percentage')
plt.title('Home vs Away Win Percentage by Team')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#teams that perform better away 
better_away = home_away[home_away['away_win_percentage'] > home_away['home_win_percentage']]

plt.figure(figsize=(8,5))
sns.barplot(x='away_win_percentage', y='team', data=better_away, color='orange')
plt.title('Teams with Better Away Win % Than Home')
plt.xlabel('Away Win %')
plt.ylabel('team')
plt.tight_layout()
plt.show()
```
